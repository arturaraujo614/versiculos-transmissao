```python
def limitar_palavras(texto):
    palavras = texto.split()
    qntd_words =len(palavras)
    lista = []

    if qntd_words <=10:
      palavras_limitadas = palavras[:qntd_words]
      texto_limitado = " ".join(palavras_limitadas)
      lista.append(texto_limitado)
      return lista

    elif qntd_words <=20:
      palavras_limitadas = palavras[:8]
      texto_limitado = " ".join(palavras_limitadas)
      lista.append(texto_limitado)

      palavras_limitadas = palavras[8:]
      texto_limitado = " ".join(palavras_limitadas)
      lista.append(texto_limitado)
      return lista

    elif qntd_words <=30:
      palavras_limitadas = palavras[:8]
      texto_limitado = " ".join(palavras_limitadas)
      lista.append(texto_limitado)

      palavras_limitadas = palavras[8:16]
      texto_limitado = " ".join(palavras_limitadas)
      lista.append(texto_limitado)

      palavras_limitadas = palavras[16:]
      texto_limitado = " ".join(palavras_limitadas)
      lista.append(texto_limitado)
      return lista

    elif qntd_words <=40:
      palavras_limitadas = palavras[:8]
      texto_limitado = " ".join(palavras_limitadas)
      lista.append(texto_limitado)

      palavras_limitadas = palavras[8:16]
      texto_limitado = " ".join(palavras_limitadas)
      lista.append(texto_limitado)

      palavras_limitadas = palavras[16:24]
      texto_limitado = " ".join(palavras_limitadas)
      lista.append(texto_limitado)

      palavras_limitadas = palavras[24:32]
      texto_limitado = " ".join(palavras_limitadas)
      lista.append(texto_limitado)

      palavras_limitadas = palavras[32:]
      texto_limitado = " ".join(palavras_limitadas)
      lista.append(texto_limitado)
      return lista

    elif qntd_words <=50:
      palavras_limitadas = palavras[:8]
      texto_limitado = " ".join(palavras_limitadas)
      lista.append(texto_limitado)

      palavras_limitadas = palavras[8:16]
      texto_limitado = " ".join(palavras_limitadas)
      lista.append(texto_limitado)

      palavras_limitadas = palavras[16:24]
      texto_limitado = " ".join(palavras_limitadas)
      lista.append(texto_limitado)

      palavras_limitadas = palavras[24:32]
      texto_limitado = " ".join(palavras_limitadas)
      lista.append(texto_limitado)

      palavras_limitadas = palavras[32:]
      texto_limitado = " ".join(palavras_limitadas)
      lista.append(texto_limitado)

      palavras_limitadas = palavras[32:40]
      texto_limitado = " ".join(palavras_limitadas)
      lista.append(texto_limitado)

      palavras_limitadas = palavras[40:]
      texto_limitado = " ".join(palavras_limitadas)
      lista.append(texto_limitado)
      return lista
```

```python
versiculos_corrigidos =[]
for x in lista_principal:
  lista_auxiliar = []
  print('---')
  for y in x:
    print(y)
    versiculo = limitar_palavras(y)
    lista_auxiliar.append(versiculo)
  print(lista_auxiliar)
  versiculos_corrigidos.append(lista_auxiliar)
```

```python
os.makedirs("Versículos Pregação", exist_ok=True)

imagem = Image.open('/content/drive/MyDrive/1.Python/Transmissao/bg.png')

# Configurar as dimensões e a posição do DataFrame
# posicao_x = 100
posicao_y = 200

# Configurar a fonte para o DataFrame
fonte = ImageFont.truetype("/content/drive/MyDrive/1.Python/IA - Academy/Reconhecimento de Texto/Fontes/arial.ttf", 80)


for k,z in enumerate(versiculos_corrigidos):
  # Criar uma nova imagem com o DataFrame sobreposto
  imagem_sobreposta = imagem.copy()
  desenho = ImageDraw.Draw(imagem_sobreposta)

# Escrever cada valor do DataFrame na imagem
  print('Iniciando o versículo:',z[-1][0])
  for i,string_resultante in enumerate(z):
    print(string_resultante[0])
    for t in string_resultante:
      # Obter o tamanho do texto
      tamanho_texto = desenho.textsize(t, font=fonte)
      # Calcular a posição x para centralizar o texto
      posicao_x_centralizado = (imagem_sobreposta.width - tamanho_texto[0]) // 2
      # Escrever o texto na imagem centralizado
      desenho.text((posicao_x_centralizado, posicao_y + i*100), t, font=fonte, fill=(255, 255, 255))

  # Obter o tamanho do texto
  tamanho_texto = desenho.textsize(z[-1][0], font=fonte)
  # Calcular a posição x para centralizar o texto
  posicao_x_centralizado = (imagem_sobreposta.width - tamanho_texto[0]) // 2
  # Escrever o texto na imagem centralizado
  desenho.text((posicao_x_centralizado, posicao_y + (i+1)*100), z[-1][0], font=fonte, fill=(255, 255, 255))

   # Gerar o nome único para a imagem
  nome_imagem = f"{k+1}.png"

    # Salvar a imagem na pasta imagens_sobrepostas
  caminho_imagem = os.path.join("Versículos Pregação", nome_imagem)
  imagem_sobreposta.save(caminho_imagem)


# Nome do arquivo ZIP a ser criado
nome_arquivo_zip = "Versículos Pregação.zip"

# Pasta que contém as imagens sobrepostas
pasta_imagens = "Versículos Pregação"

# Criar um arquivo ZIP e adicionar as imagens à ele
with zipfile.ZipFile(nome_arquivo_zip, "w") as zipf:
# Percorrer todas as imagens na pasta e adicioná-las ao arquivo ZIP
  for nome_imagem in os.listdir(pasta_imagens):
    caminho_imagem = os.path.join(pasta_imagens, nome_imagem)
    zipf.write(caminho_imagem, nome_imagem)

print("Arquivo ZIP criado com sucesso!")

# # Fazer o download do arquivo ZIP
# files.download(nome_arquivo_zip)

```
